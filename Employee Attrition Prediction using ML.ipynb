### Data Cleaning

```python
location_dict = data['Location'].value_counts()
location_dict
```

```python
location_new_dict = {
    'Chennai' :     1, 
    'Noida' :       2,
    'Bangalore' :   3, 
    'Hyderabad' :   4, 
    'Pune' :        5, 
    'Madural' :     6,
    'Lucknow' :     7,
    'Other Place' : 0
}

print(location_new_dict)
```

```python
def location(x):
    if str(x) in location_new_dict:
        return location_new_dict[str(x)]
    else:
        return location_new_dict['Other Place']

data_1 = data['Location'].apply(location)
data['New Location'] = data_1
data
```

```python
func = pd.get_dummies(data['Function'])
func
```

```python
hr = pd.get_dummies(data['Hiring Source'])
hr
```

```python
Marital_dict = data['Marital Status'].value_counts()
Marital_dict
```

```python
def Mar(x):
    if str(x) in Marital_dict.keys() and Marital_dict[str(x)] > 100:
        return str(x)
    else:
        return 'Other status'

data_1 = data['Marital Status'].apply(Mar)
data['New Marital'] = data_1
data
```

```python
MS  = pd.get_dummies(data['New Marital'])
MS.head()
```

```python
def Promoted(x):
    if x == 'Promoted':
        return int(1)
    else:
        return int(0)
    
data_3 = data['Promoted/Non Promoted'].apply(Promoted)
data['New Promotion'] = data_3
data.head()
```

```python
Emp_dict = data['Emp. Group'].value_counts()
Emp_dict
```

```python
Emp_new_dict = {
    'B1': 1,
    'B2': 2, 
    'B3': 3,
    'Other Group': 4
}
print(Emp_new_dict)
```

```python
def Emp(x):
    if str(x) in Emp_new_dict.keys():
        return str(x)
    else:
        return 'Other Group'

data_4 = data['Emp. Group'].apply(Emp)
data['New Emp Group'] = data_4

emp = pd.get_dummies(data['New Emp Group'])
data.head()
```

```python
def Job(x):
    if x == 'Yes':
        return int(1)
    else:
        return int(0)

data_5 = data['Job Role Match'].apply(Job)
data['New Job Role Match'] = data_5
data.head()
```

```python
gender_dict = data['Gender '].value_counts()
gender_dict
```

```python
gend = pd.get_dummies(data['Gender '])
gend.head()
```

```python
tengrp = pd.get_dummies(data['Tenure Grp.'])
tengrp.head()
```

```python
dataset = pd.concat([data, func, hr, MS, emp, gend, tengrp], axis = 1)
dataset.head()
```

### Model development applying 7-ML algorithms

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import datasets
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 4)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.svm import SVC
```

```python
lr = LogisticRegression(C = 0.1, random_state = 42, solver = 'liblinear')
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
gnb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors = 3)
svm = SVC(kernel = 'linear')
```

```python
for a, b in zip([lr, dt, rf, gnb, knn, svm], ['Logistic Regression', 'Decision Tree', 'Random Forest', 'Naive Bayes', 'KNN', 'SVM']):
    a.fit(x_train, y_train)
    prediction = a.predict(x_train)
    y_pred = a.predict(x_test)
    score1 = accuracy_score(y_train, prediction)
    score = accuracy_score(y_test, y_pred)
    msg1 = "[%s] training data accuracy is : %f" % (b, score1)
    msg2 = "[%s] test data accuracy is : %f" % (b, score)
    print(msg1)
    print(msg2)
```

```python
model_scores = {
    'Logistic Regression':lr.score(x_test, y_test),
    'KNN Classifier': knn.score(x_test, y_test),
    'Support Vector Machines': svm.score(x_test, y_test),
    'Random Forest': rf.score(x_test, y_test),
    'Decision Tree': dt.score(x_test, y_test),
    'Naive Bayes': gnb.score(x_test, y_test)
}
model_scores
```

```python
from sklearn.metrics import classification_report
rf_y_preds = rf.predict(x_test)
print(classification_report(y_test, rf_y_preds))
```

```python
lr_y_preds = lr.predict(x_test)
print(classification_report(y_test, lr_y_preds))
```
